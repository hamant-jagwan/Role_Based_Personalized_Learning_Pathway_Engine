## Filtering the IT Job form the Titles

```python
# Keywords covering ALL IT job roles
it_keywords = [
    # Data roles
    "data scientist", "data analyst", "data engineer", "data science",
    "business intelligence", "bi analyst", "bi developer", "analytics",
    # ML / AI
    "machine learning", "ml engineer", "deep learning",
    "ai engineer", "nlp", "natural language", "computer vision",
    # Software development
    "software engineer", "software developer", "software development",
    "python developer", "java developer", "php developer",
    "dot net developer", ".net developer", "asp.net",
    "node developer", "react developer", "angular developer",
    "wordpress developer", "magento developer", "salesforce developer",
    # Web
    "frontend developer", "front end developer", "front-end developer",
    "backend developer", "back end developer", "back-end developer",
    "full stack developer", "full-stack developer", "web developer",
    "ui developer", "web designer", "ui ux", "ui/ux",
    # Mobile
    "android developer", "ios developer", "mobile application",
    "mobile developer", "flutter developer", "react native",
    # DevOps / Cloud
    "devops", "cloud engineer", "aws engineer", "azure engineer",
    "site reliability", "sre", "infrastructure engineer",
    # Security
    "cybersecurity", "security analyst", "information security",
    "network security", "ethical hacker", "penetration tester",
    # Database
    "database administrator", "dba", "database engineer",
    # QA / Testing
    "qa engineer", "quality assurance", "test engineer",
    "automation test", "manual testing", "software tester",
    # Network / Systems
    "network engineer", "system administrator", "sysadmin",
    "network administrator", "it support", "technical support",
    # Other IT
    "business analyst", "solution architect", "technical architect",
    "tech lead", "technical lead", "scrum master", "project manager",
    "embedded", "blockchain", "game developer",
]

pattern = "|".join(it_keywords)
df = df[df["Job Title"].str.lower().str.contains(pattern, na = False)].copy()

```

```python
print(f"IT relevant rows: {len(df)}")
print("\nTop 30 Job Titles:")
print(df["Job Title"].value_counts().head(30))
```

## Mapping of Standard IT Roles

```python
# Maps every job title variation → one clean standard role name
# These role names must match exactly what is in your Streamlit dropdown

def map_to_standard_role(title):
    t = title.lower()

    # ── Data & AI ──
    if any(x in t for x in ["data scientist", "data science"]):
        return "Data Scientist"
    elif any(x in t for x in ["machine learning", "ml engineer", "deep learning", "ai engineer", "artificial intelligence"]):
        return "Machine Learning Engineer"
    elif any(x in t for x in ["data analyst", "analytics analyst"]):
        return "Data Analyst"
    elif "data engineer" in t:
        return "Data Engineer"
    elif any(x in t for x in ["business intelligence", "bi analyst", "bi developer"]):
        return "Business Intelligence Analyst"
    elif any(x in t for x in ["nlp", "natural language"]):
        return "NLP Engineer"
    elif any(x in t for x in ["computer vision", "cv engineer"]):
        return "Computer Vision Engineer"

    # ── Web & Software ──
    elif any(x in t for x in ["full stack", "full-stack"]):
        return "Full Stack Developer"
    elif any(x in t for x in ["frontend", "front end", "front-end", "ui developer", "react developer", "angular developer", "vue developer"]):
        return "Frontend Developer"
    elif any(x in t for x in ["backend", "back end", "back-end", "node developer", "django developer", "flask developer"]):
        return "Backend Developer"
    elif any(x in t for x in ["software engineer", "software developer", "software development", "python developer", "java developer",
                               "php developer", ".net developer", "dot net", "asp.net", "wordpress developer",
                               "magento developer", "salesforce developer"]):
        return "Software Engineer"

    # ── Mobile ──
    elif any(x in t for x in ["android developer", "android engineer"]):
        return "Android Developer"
    elif any(x in t for x in ["ios developer", "ios engineer", "swift developer"]):
        return "iOS Developer"
    elif any(x in t for x in ["mobile application", "flutter", "react native", "mobile developer"]):
        return "Mobile App Developer"

    # ── DevOps / Cloud ──
    elif any(x in t for x in ["devops", "site reliability", "sre"]):
        return "DevOps Engineer"
    elif any(x in t for x in ["cloud engineer", "aws engineer", "azure engineer", "gcp engineer", "infrastructure engineer"]):
        return "Cloud Engineer"

    # ── Security ──
    elif any(x in t for x in ["cybersecurity", "security analyst", "information security", "network security", "ethical hacker", "penetration"]):
        return "Cybersecurity Analyst"

    # ── Database ──
    elif any(x in t for x in ["database administrator", "dba", "database engineer"]):
        return "Database Administrator"

    # ── QA ──
    elif any(x in t for x in ["qa engineer", "quality assurance", "test engineer", "automation test", "software tester"]):
        return "QA Engineer"

    # ── Network / Systems ──
    elif any(x in t for x in ["network engineer", "network administrator"]):
        return "Network Engineer"
    elif any(x in t for x in ["system administrator", "sysadmin", "it support", "technical support"]):
        return "System Administrator"

    # ── UI/UX ──
    elif any(x in t for x in ["ui ux", "ui/ux", "ux designer", "web designer", "product designer"]):
        return "UI/UX Designer"

    # ── Other IT ──
    elif any(x in t for x in ["business analyst"]):
        return "Business Analyst"
    elif any(x in t for x in ["solution architect", "technical architect"]):
        return "Solutions Architect"
    elif any(x in t for x in ["tech lead", "technical lead"]):
        return "Tech Lead"
    elif any(x in t for x in ["embedded"]):
        return "Embedded Systems Engineer"
    elif "blockchain" in t:
        return "Blockchain Developer"
    else:
        return None
```

```python
df["role_name"] = df["Job Title"].apply(map_to_standard_role)
df = df.dropna(subset=["role_name"])

print(f"Rows after role mapping: {len(df)}")
print("\nRole distribution:")
print(df["role_name"].value_counts())
```

## Clean & Normalize Skill Names

```python
def clean_skills(skill_string):
    skills = skill_string.split("|")
    skills  = [s.strip().lower() for s in skills if s.strip()]
    return skills

df["skills_list"] = df["Key Skills"].apply(clean_skills)
print(df["skills_list"].head(2))
```

```python
# Normalize abbreviations to full names
synonym_map = {
    "ml"          : "machine learning",
    "ai"          : "artificial intelligence",
    "dl"          : "deep learning",
    "nlp"         : "natural language processing",
    "cv"          : "computer vision",
    "dbms"        : "database management",
    "ds"          : "data science",
    "stats"       : "statistics",
    "tf"          : "tensorflow",
    "sklearn"     : "scikit-learn",
    "js"          : "javascript",
    "ts"          : "typescript",
    "k8s"         : "kubernetes",
    "oop"         : "object oriented programming",
    "oops"        : "object oriented programming",
    "dsa"         : "data structures and algorithms",
    "ci/cd"       : "ci cd pipelines",
    "restapi"     : "rest api",
    "rest apis"   : "rest api",
}

def apply_synonyms(skills):
    return [synonym_map.get(s, s) for s in skills]

df["skills_list"] = df["skills_list"].apply(apply_synonyms)

print("Sample cleaned skills:")
for row in df["skills_list"].head(5).tolist():
    print(" ", row)
```

## Calculate Importance Score
`importance_score = skill_count / total_jobs_for_that_role`

```python
df.head(3)
```

```python
# Step 1: Total jobs per role
total_jobs_per_role = df.groupby("role_name").size().reset_index(name = "total_jobs")
print("Total jobs per role:")
print(total_jobs_per_role.to_string(index = False))
```

```python
# Step 2: Explode — one row per skill per job
df_exploded = df[["role_name", "skills_list"]].explode("skills_list")
df_exploded = df_exploded.rename(columns={"skills_list": "required_skill"})
df_exploded = df_exploded[df_exploded["required_skill"].str.len() > 2]
print(df_exploded)
```

```python
# Step 3: Count skill frequency per role
skill_counts = df_exploded.groupby(
    ["role_name", "required_skill"]
).size().reset_index(name="skill_count")
print(skill_counts)
```

```python
# Step 4: Merge with total jobs
result = skill_counts.merge(total_jobs_per_role, on="role_name")
print(result)
```

```python
# Step 5: Calculate importance score
result["importance_score"] = (result["skill_count"] / result["total_jobs"]).round(2)

print(result["importance_score"])
```

```python
# Step 6: Keep skills appearing in at least 5% of jobs for that role
result = result[result["importance_score"] >= 0.05]
print(result)
```

```python
# Step 7: Sort by role then importance descending
result = result.sort_values(
    ["role_name", "importance_score"],
    ascending=[True, False]
).reset_index(drop=True)

print(f"\nTotal role-skill pairs: {len(result)}")
print("\nSample:")
print(result.head(20).to_string(index=False))
```

## Quality Check
